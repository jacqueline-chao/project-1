```python
import pandas as pd
import numpy as np
#loading in the data
file_path = '/sub-data.txt'
df = pd.read_csv(file_path, delimiter=',')
df.head()
```

```python
print(df.shape, '\n') #4032 rows and 7 columns
```

From the codebook, '.d'= 'don't know', '.n'='no answer', '.i'= 'iap' (inapplicable), and '.s'='skipped on web'."Inapplicable" means that the question was not included as part of the survey for some respondents.

```python
#checking value of 'age' variable
var = 'age'
print(df[var].unique(),'\n') # need to check out '.n' and '.i'
print(df[var].value_counts(), '\n')
```

```python
#turn .n and .i into NaNs for age variable
#df[var] = pd.to_numeric(df[var], errors='coerce') # Coerce the variable to numeric
df[var] = df[var].replace('.n',np.nan) # Notice the column replacement
df[var] = df[var].replace('.i',np.nan)
print(df[var].unique(), '\n')
print(df[var].value_counts(), '\n')
print(df[var].describe(),'\n') # Describe the numeric variable
#df[var].hist(bins = 50) # Histogram of the variable values
#print('Total Missings: \n', sum(df[var+'_nan']),'\n') # How many missing values are there?
```

```python
#checking value of 'conmedic' variable
var = 'conmedic'
print(df[var].unique(),'\n') # need to check out '.i', '.s', and '.d'
#don't know is probably the same as skipped on web (if you don't know, you're likely to just skip the question)
#or could be skipped if it's further down in the survey
print(df[var].value_counts(), '\n')
```

```python
#turn .i,.s, and .d into NaNs for conmedic variable
var='conmedic'
df[var] = df[var].replace('.i',np.nan) # Notice the column replacement
df[var] = df[var].replace('.s',np.nan)
df[var] = df[var].replace('.d',np.nan)
print(df[var].unique(), '\n')
print(df[var].value_counts(), '\n')
print(df[var].describe(),'\n')
```

```python
#checking value of 'region' variable
var = 'region'
print(df[var].unique(),'\n')
print(df[var].value_counts(), '\n')
#'region' has no missing values
```

```python
#checking value of 'partyid' variable
var = 'partyid'
print(df[var].unique(),'\n') # need to check out '.n'
print(df[var].value_counts(), '\n')
```

```python
#turn .n into NaN for partyid variable
var='partyid'
df[var] = df[var].replace('.n',np.nan)
print(df[var].unique(), '\n')
print(df[var].value_counts(), '\n')
print(df[var].describe(),'\n')
```

```python
#checking value of 'race' variable
var = 'race'
print(df[var].unique(),'\n') # need to check out '.i'
print(df[var].value_counts(), '\n')
```

```python
#turn .i into NaN for race variable
var='race'
df[var] = df[var].replace('.i',np.nan)
print(df[var].unique(), '\n')
print(df[var].value_counts(), '\n')
print(df[var].describe(),'\n')
```

```python
#checking value of 'sex' variable
var = 'sex'
print(df[var].unique(),'\n') # need to check out '.n', '.i', and '.s'
print(df[var].value_counts(), '\n')
```

```python
#turn .n into NaN for sex variable
var='sex'
df[var] = df[var].replace('.n',np.nan)
df[var] = df[var].replace('.i',np.nan)
df[var] = df[var].replace('.s',np.nan)
print(df[var].unique(), '\n')
print(df[var].value_counts(), '\n')
print(df[var].describe(),'\n')
```

```python
#Rename columns
df.rename(columns={'age': 'Age'}, inplace=True)
df.rename(columns={'region': 'Region'}, inplace=True)
df.rename(columns={'race': 'Race'}, inplace=True)
df.rename(columns={'sex': 'Sex'}, inplace=True)
df.rename(columns={'year': 'Year'}, inplace=True)
df.rename(columns={'conmedic': 'Confidence in Medicine'}, inplace=True)
df.rename(columns={'partyid': 'Political Party'}, inplace=True)
df.head()
```

```python
# Use dtypes attribute to get data types of each column
column_types = df.dtypes
print(column_types)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
print(df.dtypes)
# Fill NaN values with a placeholder value (e.g., 'Missing')
df['Confidence in Medicine'].fillna('Missing', inplace=True)
# Convert to categorical
df['Confidence in Medicine'] = pd.Categorical(df['Confidence in Medicine'])
# Now,is a categorical column with NaN preserved as a category
print(df['Confidence in Medicine'])
#do same thing for race
# Fill NaN values with a placeholder value (e.g., 'Missing')
df['Race'].fillna('Missing', inplace=True)
# Convert to categorical
df['Race'] = pd.Categorical(df['Race'])
# Now,is a categorical column with NaN preserved as a category
print(df['Race'])
#same for sex
# Fill NaN values with a placeholder value (e.g., 'Missing')
df['Sex'].fillna('Missing', inplace=True)
# Convert to categorical
df['Sex'] = pd.Categorical(df['Sex'])
# Now,is a categorical column with NaN preserved as a category
print(df['Sex'])
#do same for political party
# Fill NaN values with a placeholder value (e.g., 'Missing')
df['Political Party'].fillna('Missing', inplace=True)
# Convert to categorical
df['Political Party'] = pd.Categorical(df['Political Party'])
# Now,is a categorical column with NaN preserved as a category
print(df['Political Party'])

#convert region to categorical
df['Region'] = pd.Categorical(df['Region'])

#see the datatypes for all of the columns
print(df.dtypes)
```

```python
#create new csv file for clean data
df.to_csv("GSS_data_2021_clean", index=False)
#viewing the new csv file for the cleaned 2021 GSS data
file_path = "/content/GSS_data_2021_clean"
clean_2021_data = pd.read_csv(file_path, delimiter=',')
clean_2021_data.head()
```
